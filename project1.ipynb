```python
import pandas as pd

# Load the dataset
file_path = "Salary_Dataset_with_Extra_Features.csv"
df = pd.read_csv(file_path)

# Display basic info about the dataset
print("Dataset Information:")
print(df.info())

# Display first few rows
print("\nFirst 5 Rows:")
print(df.head())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Fill missing values in Salary with median salary
df['Salary'].fillna(df['Salary'].median(), inplace=True)

# Drop rows where Company Name, Job Title, or Location is missing
df.dropna(subset=['Company Name', 'Job Title', 'Location'], inplace=True)

# Ensure Salary is numeric
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

# Summary statistics for salary
print("\nSalary Summary:")
print(df['Salary'].describe())

# Grouping by Company Name and getting average salary
company_salary = df.groupby('Company Name')['Salary'].mean().reset_index()
print("\nAverage Salary by Company:")
print(company_salary)

# Grouping by Job Title and getting average salary
job_salary = df.groupby('Job Title')['Salary'].mean().reset_index()
print("\nAverage Salary by Job Title:")
print(job_salary)

# Grouping by Location and getting average salary
location_salary = df.groupby('Location')['Salary'].mean().reset_index()
print("\nAverage Salary by Location:")
print(location_salary)

# Grouping by Employment Status and getting average salary
employment_salary = df.groupby('Employment Status')['Salary'].mean().reset_index()
print("\nAverage Salary by Employment Status:")
print(employment_salary)

# Display unique job roles
print("\nUnique Job Roles:")
print(df['Job Roles'].unique())

# Highest and lowest salaries
print("\nTop 5 Highest Salaries:")
print(df.nlargest(5, 'Salary'))

print("\nTop 5 Lowest Salaries:")
print(df.nsmallest(5, 'Salary'))

# Top-paying companies
top_companies = df.groupby('Company Name')['Salary'].mean().nlargest(10)
print("\nTop 10 Highest Paying Companies:")
print(top_companies)

# Most common job roles
print("\nMost Common Job Roles:")
print(df['Job Roles'].value_counts().head(10))

# Salary comparison for full-time vs. part-time
employment_salary_comparison = df.groupby('Employment Status')['Salary'].mean()
print("\nSalary Comparison (Full-Time vs. Part-Time):")
print(employment_salary_comparison)
```

```python
print("\nChecking Salary Outliers:")
print(df['Salary'].describe())  # Check min, max, mean, etc.

# Removing extreme outliers (e.g., salaries above the 99th percentile)
upper_limit = df['Salary'].quantile(0.99)
df = df[df['Salary'] <= upper_limit]

print("\nAfter Outlier Removal:")
print(df['Salary'].describe())
```

```python
print("\nTop 10 Highest Paying Job Titles:")
print(df.groupby('Job Title')['Salary'].mean().nlargest(10))

print("\nTop 10 Lowest Paying Job Titles:")
print(df.groupby('Job Title')['Salary'].mean().nsmallest(10))
```

```python
print("\nSalary by Employment Status:")
print(df.groupby('Employment Status')['Salary'].mean())
```

```python
df.to_csv("Final_Salary_Dataset.csv", index=False)
print("\nCleaned and Processed Dataset Saved Successfully!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (assuming it's already cleaned and processed)
file_path = "Final_Salary_Dataset.csv"
df = pd.read_csv(file_path)

# Ensure 'Salary' is numeric
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')

# Set Seaborn style and context
sns.set_style("whitegrid")
sns.set_context("talk")

# ------------------------------
# MATPLOTLIB VISUALIZATION TASKS
# ------------------------------

# 1. Simple Line Plot: Top 10 Job Titles by Average Salary
job_salary = df.groupby("Job Title")["Salary"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.plot(job_salary.index, job_salary.values, marker='o', linestyle='-', color='b')
plt.xlabel("Job Title")
plt.ylabel("Average Salary (INR)")
plt.title("Top 10 Job Titles by Average Salary")
plt.xticks(rotation=45, ha="right")
plt.grid(True)
plt.legend(["Avg Salary"])
plt.tight_layout()
plt.show()

# 2. Scatter Plot: Salary vs. Salaries Reported
plt.figure(figsize=(10, 6))
plt.scatter(df["Salary"], df["Salaries Reported"], color='green', marker='x')
plt.xlabel("Salary (INR)")
plt.ylabel("Salaries Reported")
plt.title("Scatter Plot: Salary vs. Salaries Reported")
plt.grid(True)
plt.tight_layout()
plt.show()

# 3. Histogram: Distribution of Salary
plt.figure(figsize=(10, 6))
plt.hist(df["Salary"].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.xlabel("Salary (INR)")
plt.ylabel("Frequency")
plt.title("Histogram of Salary Distribution")
plt.tight_layout()
plt.show()

# 4. Bar Chart: Average Salary by Employment Status
emp_salary = df.groupby("Employment Status")["Salary"].mean().reset_index()
plt.figure(figsize=(8, 5))
plt.bar(emp_salary["Employment Status"], emp_salary["Salary"], color=['orange', 'blue', 'green', 'red'])
plt.xlabel("Employment Status")
plt.ylabel("Average Salary (INR)")
plt.title("Average Salary by Employment Status")
plt.tight_layout()
plt.show()

# -----------------------------
# SEABORN VISUALIZATION TASKS
# -----------------------------

# 1. Box Plot: Salary by Top 10 Job Titles to Identify Outliers
top_job_titles = df["Job Title"].value_counts().head(10).index
plt.figure(figsize=(12, 6))
sns.boxplot(x="Salary", y="Job Title", data=df[df["Job Title"].isin(top_job_titles)], palette="Set3")
plt.title("Box Plot of Salary by Top 10 Job Titles")
plt.xlabel("Salary (INR)")
plt.ylabel("Job Title")
plt.tight_layout()
plt.show()

# 2. Heatmap: Correlation Matrix of Numerical Features
plt.figure(figsize=(8, 6))
numerical_cols = ["Rating", "Salary", "Salaries Reported"]
corr_matrix = df[numerical_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.tight_layout()
plt.show()

# 3. Violin Plot: Salary Distribution by Employment Status
plt.figure(figsize=(8, 6))
sns.violinplot(x="Employment Status", y="Salary", data=df, palette="muted")
plt.title("Violin Plot of Salary Distribution by Employment Status")
plt.xlabel("Employment Status")
plt.ylabel("Salary (INR)")
plt.tight_layout()
plt.show()

# 4. Pair Plot: Exploring Relationships Among Numerical Features
sns.pairplot(df[numerical_cols])
plt.suptitle("Pair Plot of Numerical Features", y=1.02)
plt.tight_layout()
plt.show()

# 5. Seaborn Bar Plot: Top 10 Locations by Average Salary
location_salary = df.groupby("Location")["Salary"].mean().sort_values(ascending=False).head(10).reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x="Location", y="Salary", data=location_salary, palette="viridis")
plt.title("Top 10 Locations by Average Salary")
plt.xlabel("Location")
plt.ylabel("Average Salary (INR)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 6. Swarm Plot: Distribution of Salary by Employment Status
# Sample the data to avoid performance issues (e.g., use 1000 random rows)
df_sample = df.sample(n=1000, random_state=42)
plt.figure(figsize=(10, 6))
sns.swarmplot(x="Employment Status", y="Salary", data=df_sample, palette="deep")
plt.title("Swarm Plot of Salary by Employment Status")
plt.xlabel("Employment Status")
plt.ylabel("Salary (INR)")
plt.tight_layout()
plt.show()

# 7. KDE Plot: Density of Salary Distribution
plt.figure(figsize=(10, 6))
sns.kdeplot(df["Salary"].dropna(), shade=True, color="r")
plt.title("KDE Plot of Salary Distribution")
plt.xlabel("Salary (INR)")
plt.ylabel("Density")
plt.tight_layout()
plt.show()

# 8. Scatter Plot with Hue: Salary vs. Rating, Colored by Employment Status
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Rating", y="Salary", hue="Employment Status", data=df, palette="bright")
plt.title("Scatter Plot: Salary vs. Rating by Employment Status")
plt.xlabel("Rating")
plt.ylabel("Salary (INR)")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define points
x = [0, 1, 1, 0, 0.5]
y = [0, 0, 1, 1, 1.5]
# Define connections
edges = [(0, 1), (1, 2), (2, 3), (3, 0), (2, 4), (3, 4)]
# Plot edges
plt.figure(figsize=(5, 5))
for edge in edges:
    x_vals = [x[edge[0]], x[edge[1]]]
    y_vals = [y[edge[0]], y[edge[1]]]
    plt.plot(x_vals, y_vals, 'k-')
# Plot nodes
plt.scatter(x, y, color='lightblue', s=200)
plt.show()
```
